# import packages

```python

###1.read the data 
head_df=pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Regression/headbrain.csv")
```

```python
##3.separate the input and output
input_data=head_df.drop('Brain Weight(grams)',axis=1)
output_data=head_df['Brain Weight(grams)']
input_data.head()
```

```python
input_data.columns
```

```python
###4.applying the model
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
```

```python
###5.train the model by using fit method
linear.fit(input_data,output_data)
```

```python
##6.test the model by using  predict method
pred=linear.predict(input_data)
pred
```

```python
###evalute the model
linear.score(input_data,output_data)*100
```

```python
apple_df=pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Regression/AppleStore.csv")
apple_df.head()
```

```python
apple_df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lbr=LabelEncoder()
```

```python
apple_df['currency']=lbr.fit_transform(apple_df['currency'])
apple_df['ver']=lbr.fit_transform(apple_df['ver'])
apple_df['cont_rating']=lbr.fit_transform(apple_df['cont_rating'])
apple_df['prime_genre']=lbr.fit_transform(apple_df['prime_genre'])
```

```python
apple_df.columns
```

```python
apple_df.drop(['Unnamed: 0','track_name'],axis=1,inplace=True)
apple_df.head()
```

```python
apple_df.head()
```

```python
apple_df.columns
```

```python
x=apple_df.drop('price',axis=1)
y=apple_df['price']
```

```python
import seaborn as sns
sns.pairplot(apple_df)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
y_pred=lr.predict(x)
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_pred)*100
```
